# Merging Dataframes
#### 1. Merging is about joining two different data frame by a common attribute
#### 2. In SQL this is called Join
#### 3. As in SQL we have outer, inner, left inner and right inner joins
#### 4. These are anologous to Union, Intersection and Set A intercsection B and Set B intersection A

```python
# We are creating a data frame here
# Values passed as dictiony for DF where key becomes column name and value becomes the respective value of it
# Index are generally auto generated unless it is explicitly specified
df = pd.DataFrame([{'Name': 'Chris', 'Item Purchased': 'Sponge', 'Cost': 22.50},
                   {'Name': 'Kevyn', 'Item Purchased': 'Kitty Litter', 'Cost': 2.50},
                   {'Name': 'Filip', 'Item Purchased': 'Spoon', 'Cost': 5.00}],
                  index=['Store 1', 'Store 1', 'Store 2'])
df
```

```python
# Appending a new column in Pandas data frame
df['Date'] = ['December 1', 'January 1', 'mid-May']
df
```

```python
# Appending a boolean value to the DF, this is called general assignment, value is applied to all rows in a DF
df['Delivered'] = True
df
```

```python
# We can inject None value too for any particular cell
df['Feedback'] = ['Positive', None, 'Negative']
df
```

```python
# we can reindex a column using reset_index(), this will remove the existing index and apply default numerical index
adf = df.reset_index()
adf['Date'] = pd.Series({0: 'December 1', 2: 'mid-May'})
adf
```

```python
# creating two difference data frame 
staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR'},
                         {'Name': 'Sally', 'Role': 'Course liasion'},
                         {'Name': 'James', 'Role': 'Grader'}])
staff_df = staff_df.set_index('Name')

student_df = pd.DataFrame([{'Name': 'James', 'School': 'Business'},
                           {'Name': 'Mike', 'School': 'Law'},
                           {'Name': 'Sally', 'School': 'Engineering'}])
student_df = student_df.set_index('Name')

staff_df.head(3)
```

```python
student_df.head(3)
```

```python
# outer join - This is union in set language
# if we need to merge based on index then we used left_index and right_index - this is joining based on a common parameter
pd.merge(staff_df, student_df, how='outer', left_index=True, right_index=True)
```

```python
# this is called as intersection
# inner join in SQL
# only those values common in both set A an set B will be displayed
pd.merge(staff_df, student_df, how='inner', left_index=True, right_index=True)
```

```python
# left inner join takes all values in set A with its respective values in set B
# this is set A given B
pd.merge(staff_df, student_df, how='left', left_index=True, right_index=True)
```

```python
# Right inner join takes all values in set B with its respective values in set A
# this is set B given A
pd.merge(staff_df, student_df, how='right', left_index=True, right_index=True)
```

```python
# left_on and right_on is used to jon two DF using a particular column instead of index
# Join will be based on columns instead of row index
staff_df = staff_df.reset_index()
student_df = student_df.reset_index()
pd.merge(staff_df, student_df, how='left', left_on='Name', right_on='Name')
```

```python
# When we merge two dataset which has same column names one on left wil be referred with _x and one on left will be referred _y
# We merge here based on Name. However, location column is common on two DF
# So, when we merger for Staff_df location will be refered as Location_x, while student_df's will be location_y
staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR', 'Location': 'State Street'},
                         {'Name': 'Sally', 'Role': 'Course liasion', 'Location': 'Washington Avenue'},
                         {'Name': 'James', 'Role': 'Grader', 'Location': 'Washington Avenue'}])

student_df = pd.DataFrame([{'Name': 'James', 'School': 'Business', 'Location': '1024 Billiard Avenue'},
                           {'Name': 'Mike', 'School': 'Law', 'Location': 'Fraternity House #22'},
                           {'Name': 'Sally', 'School': 'Engineering', 'Location': '512 Wilson Crescent'}])

staff_df
```

```python
student_df
```

```python
# performing left inner join and not including a comman collumn as a join parameter, new col names with _x and _y are induced
# Also see we have done a left inner join
pd.merge(staff_df, student_df, how='left', left_on='Name', right_on='Name')
```

```python
# We can even pass list to the left_on and right_on to merge based on multiple columns from two DF
# Thus data is merged only when all column value get matched
staff_df = pd.DataFrame([{'First Name': 'Kelly', 'Last Name': 'Desjardins', 'Role': 'Director of HR'},
                         {'First Name': 'Sally', 'Last Name': 'Brooks', 'Role': 'Course liasion'},
                         {'First Name': 'James', 'Last Name': 'Wilde', 'Role': 'Grader'}])
student_df = pd.DataFrame([{'First Name': 'James', 'Last Name': 'Hammond', 'School': 'Business'},
                           {'First Name': 'Mike', 'Last Name': 'Smith', 'School': 'Law'},
                           {'First Name': 'Sally', 'Last Name': 'Brooks', 'School': 'Engineering'}])
staff_df
```

```python
student_df
```

```python
# We will merge two DF using multiple column here
# only those records/tuples which are matched with all column specified between two DF will be merged
pd.merge(staff_df, student_df, how='inner', left_on=['First Name','Last Name'], right_on=['First Name','Last Name'])
```

# Idiomatic Pandas: Making Code Pandorable
#### 1. This is about best coding practice
#### 2. Efficient way of using the moduler programming
#### 3. In python modular programming yeald much faster result than a naive implementation because of vectorized computation
#### 4, This is called method chaining, where we chain result of one method as input to another method and so on 
#### 5. This a pandorable way of approaching a problem

```python
# Pandorable means writing an efficient code by ignoring the ][ in data selection
# reading a data through pandas
df = pd.read_csv('census.csv')
```

```python
# this is apandorable way of implementing method chaining
# data frame is defines once and each method is called with a period being sufficed
# this is the best way to use the python methods instead of having multiple ][ like df['state']['country']
(df.where(df['SUMLEV']==50)
    .dropna()
    .set_index(['STNAME','CTYNAME'])
    .rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'}))
```

```python
# importing data again
df = pd.read_csv('census.csv')
```

```python
# this is a naive way of performing the above method chaining, this is not pandorable
df = df[df['SUMLEV']==50]
df.dropna()
df.set_index(['STNAME','CTYNAME'], inplace=True)
df.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})
```

```python
# Pactice of class Question
#(df.drop(df[df['Quantity'] == 0].index)
# .rename(columns={'Weight': 'Weight (oz.)'})
```

```python
# takes a row as its input
# project only few columns which are required 
# then find min and max for those columns
def min_max(row):
    #print(row)
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'min': np.min(data), 'max': np.max(data)})
```

# Apply

```python
# apply is a about applying a certain function for each row iteratively
# min max operation is gonna happen based on each column in the min_max function through it is a row level operation
# so using axis=1
df.apply(min_max, axis=1).head(5)
```

```python
# adding min and max as a new column to the existing dataframe
# this is the most common approach done in most of the data analysis projects
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    row['max'] = np.max(data)
    row['min'] = np.min(data)
    return row
df.apply(min_max, axis=1).head(5)
```

```python
# pandorable way of coding the same using the lambda function
# but we can only do one anonomous operation using lambda function
rows = ['POPESTIMATE2010',
        'POPESTIMATE2011',
        'POPESTIMATE2012',
        'POPESTIMATE2013',
        'POPESTIMATE2014',
        'POPESTIMATE2015']
df.apply(lambda x: np.max(x[rows]), axis=1).head(5)
```

# Group by
#### 1. This is used to group records in a data frame
#### 2. Grouping is genrally done on a common index, that is grouping all records whose index are same
#### 3. Thus by grouping, we get a condition based on grouping done and its particulat DF
#### 4. This grouping alone is not important, it is about what we are further going to do with this grouping is important
#### 5. Aggregation of certain column is the most important operation along with grouping

```python
# this group by will take some column names as input and split the data into some chunks
# Loop + unique function = Group by
# return is a groupby object which need to be iterated to see the results
# return group by object is a tuple = (Group condition, grouped dataframe)
df = pd.read_csv('census.csv')
df = df[df['SUMLEV']==50]
df.head(5)
```

```python
%%timeit -n 10
# performing group function without groupby method - Unique + iteration dont through Unique and for loop respectively
# this operation will consume more time
for state in df['STNAME'].unique():
    avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
    print('Counties in state ' + state + ' have an average population of ' + str(avg))
```

```python
%%timeit -n 10
# this is a group by way of implementing above function in a more pandorable way
# group by takes are of grouping values in to chunk of many DF
# these DF are then iterated and operated
for group, frame in df.groupby('STNAME'):
    avg = np.average(frame['CENSUS2010POP'])
    print('Counties in state ' + group + ' have an average population of ' + str(avg))
```

```python
# let is read the data again
df = pd.read_csv('census.csv')
df = df.set_index('STNAME')
df.head(5)
```

```python
# below methid is called light weight hashing
# here we create a function and pass this function to a groupby method
# groupby method will then group records based on the outcome of return values
# split - combine - apply

# fun() function is used to take the states value, 
# check its first char and if it is less than M that is between 'A' and 'L' we will classify them as 0 and 
# anything between M and P are classified as 1 and others are classified as 2

def fun(item):
    #print(item[0])
    if item[0]<'M':
        return 0
    if item[0]<'Q':
        return 1
    return 2

# dont get confused seeing no arg passed to the fun() function
# df.groupby(fun) means we say python to validate the whole data frame, look the index and apply fun function 
for group, frame in df.groupby(fun):
    print('There are ' + str(len(frame)) + ' records in group ' + str(group) + ' for processing.')
```

# Aggregate with Group By

```python
# agg function - useful for any data manipulation based on the result obtained from a group by function
# this is purely an aggregation function to do some mathematical operation
# we pass key-value which is the label - actual agg operation for the result to be obtained 
df.groupby('STNAME').agg({'CENSUS2010POP': np.average})
```

```python
# there is a 2 different group by objects
# 1. series group by object 
# 2. Dataframe group by object
print(type(df.groupby(level=0)['POPESTIMATE2010','POPESTIMATE2011']))
print(type(df.groupby(level=0)['POPESTIMATE2010']))
```

```python
# making use of series group by object
# here we have just one column basd on which our agg function is applied to
(df.set_index('STNAME').groupby(level=0)['CENSUS2010POP']
    .agg({'avg': np.average, 'sum': np.sum}))
```

```python
# agg applied to column group by object
# it automatiacally labels the column header
# output returned itself is a data frame
(df.set_index('STNAME').groupby(level=0)['POPESTIMATE2010','POPESTIMATE2011']
    .agg({'avg': np.average, 'sum': np.sum}))
```

```python
# agg applied to dataframe group by object
# however applying one mathematical operation for one column 
# ideally pandas should not allow to use the same column name as label passed for key in dictionary but this is how it is
(df.set_index('STNAME').groupby(level=0)['POPESTIMATE2010','POPESTIMATE2011']
    .agg({'POPESTIMATE2010': np.average, 'POPESTIMATE2011': np.sum}))
```

# Scales
#### 1. This is about type casting a attribute, mainly deal with categorical
#### 2. Important scales are a. Ratio Scale, b. Interval Scale, c. Nominal Scale and d. Ordinal Scale

```python
# creating a simple DF with categorical values
df = pd.DataFrame(['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D'],
                  index=['excellent', 'excellent', 'excellent', 'good', 'good', 'good', 'ok', 'ok', 'ok', 'poor', 'poor'])
df.rename(columns={0: 'Grades'}, inplace=True)
df
```

```python
# turining a nominal value to a Ordinal value
# using astype() methos and passing categories in a certain ranking order with ordered = True 
grades = df['Grades'].astype('category',
                             categories=['D', 'D+', 'C-', 'C', 'C+', 'B-', 'B', 'B+', 'A-', 'A', 'A+'],
                             ordered=True)
grades.head()
# below you see the type as category with some order like D < D+
```

```python
# with this we can do necessary boolean masking
grades > 'C'
```

```python
# we can create bins as used in hist or discretize data as below
df = pd.read_csv('census.csv')
df = df[df['SUMLEV']==50]
df = df.set_index('STNAME').groupby(level=0)['CENSUS2010POP'].agg({'avg': np.average})
pd.cut(df['avg'],10)
```

# Pivot Tables
#### 1. This is a Data Frame where each cell consist of some aggregated values wrt corresponding rows and columns
#### 2. aggfunc is used to mention the type of arithmetic operation we need to do
#### 3. multiple arithmatic operation is done by passing them in list

```python
#http://open.canada.ca/data/en/dataset/98f1a129-f628-4ce4-b24d-6f16bf24dd64
df = pd.read_csv('cars.csv')
```

```python
df.head()
```

```python
# here we are using Year as index, and car make as columns and each cell value must contain the mean of these values
# that is in 2012 what is the mean KW value for Mitsubishi
df.pivot_table(values='(kW)', index='YEAR', columns='Make', aggfunc=np.mean)
```

```python
# passing multiple aggregation methods
df.pivot_table(values='(kW)', index='YEAR', columns='Make', aggfunc=[np.mean,np.min], margins=True)
```
